```python
import tensorflow as tf
import numpy as np
```

```python
C = np.array([7,-4,2,0,-5,2,1], dtype = float)
F = np.array([40,54,100,20,15,43,8], dtype = float)
```

```python
# Usamos Keras para añadir las capas de la RN

capa = tf.keras.layers.Dense(units=1,input_shape=[1])
```

```python
# Especificamos el modelo

modelo = tf.keras.Sequential([capa])
```

```python
# Se añade el compilador
modelo.compile(
      optimizer = tf.keras.optimizers.Adam(0.1),
      loss = 'mean_squared_error'
)
```

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(C,F,epochs = 1000,verbose = False)
print("Modelo entrenado")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
print("Hgamos una predicción")
resultado = modelo.predict([100.0])
print("El resultado es" + str(resultado) + "Farenheit!")
```
